```python
# Creamos un grafo no dirigido usando NetworkX
g = nx.Graph()

# Añadimos nodos y aristas
edges = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 5), (4, 5), (4, 6), (4, 7), (5, 6), (6, 7)]
g.add_edges_from(edges)
```

```python
# Mostramos información sobre el grafo
print("Clase del grafo:", type(g))
print("ID del grafo:", id(g))
```

```python
print("Nodos:", list(g.nodes()))
print("Cantidad de nodos:", g.number_of_nodes())
```

```python
print("Aristas:", list(g.edges()))
print("Cantidad de aristas:", g.number_of_edges())
```

```python
print(g)
```

```python
print("¿El grafo es ponderado?", any('weight' in g[u][v] for u, v in g.edges()))
print("¿El grafo es simple?", g.is_multigraph())
```

```python
pos = nx.kamada_kawai_layout(g)

plt.figure(figsize=(8, 6))
nx.draw(g, pos, with_labels=True, node_color='yellow', node_size=1000, font_size=12, font_weight='bold')
plt.title("Red binaria no dirigida")
plt.show()
```

```python
# EJEMPLO 3.2
```

```python
# Creamos una copia del grafo original para mantener la estructura del grafo no dirigido
wg = g.copy()

# Asignamos pesos aleatorios a las aristas desde una distribución continua definida por el usuario
random.seed(123)
for u, v in wg.edges():
    wg[u][v]['weight'] = round(random.uniform(0, 1), 3)
```

```python
# Mostramos los pesos asignados a las aristas
print("Pesos de las aristas:", [wg[u][v]['weight'] for u, v in wg.edges()])

# Mostramos si el grafo es ponderado
print("¿El grafo es ponderado?", any('weight' in wg[u][v] for u, v in wg.edges()))
```

```python
# Dibuja el grafo ponderado
random.seed(123)

#pos = nx.kamada_kawai_layout(wg)

plt.figure(figsize=(8, 6))
edge_widths = [5 * wg[u][v]['weight'] for u, v in wg.edges()]
nx.draw(wg, with_labels=True, node_color='yellow', node_size=1000, font_size=12, font_weight='bold', width=edge_widths)
plt.title("Red ponderada no dirigida")
plt.show()
```

```python
# Creamos un grafo dirigido usando NetworkX
dg = nx.DiGraph()

# Añadimos nodos
dg.add_node(1, name="Juan", sexo="M")
dg.add_node(2, name="Maria", sexo="F")
dg.add_node(3, name="Pedro", sexo="M")

# Añadimos aristas
dg.add_edge(1, 2)
dg.add_edge(1, 3)
dg.add_edge(2, 3)
dg.add_edge(3, 2)
```

```python
# Mostramos las aristas
print("Aristas:", list(dg.edges()))
```

```python
# Creamos un diccionario de nombres de nodos para usar como etiquetas en el gráfico
node_labels = {node: data['name'] for node, data in dg.nodes(data=True)}

# Dibujamos el grafo dirigido con etiquetas de nodos

pos = nx.kamada_kawai_layout(dg)

plt.figure(figsize=(8, 6))
nx.draw(dg, pos, with_labels=True, labels=node_labels, node_size=2000, font_size=12, font_weight='bold', node_color='yellow', arrows=True)
plt.title("Red binaria dirigida")
plt.show()
```
