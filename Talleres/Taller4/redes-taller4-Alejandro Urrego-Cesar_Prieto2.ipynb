```python
from youtube_transcript_api import YouTubeTranscriptApi
from pytube import Playlist
import re
import os
```

```python

def get_playlist(playlists):
    urls=[]
    for playlist in playlists:
        playlist_urls=Playlist(playlist)
        for url in playlist_urls:
            urls.append(url)
    return(urls)
```

```python
playlist=['https://www.youtube.com/playlist?list=PLbtegW3d3L4Id_lrAgBRDGf-k_bwEYH6_']
urls=get_playlist(playlist)
```

```python
patron = r'https://www.youtube.com/watch\?v=(.+)'
coincidencias = re.search(patron, urls[0])
id_video = coincidencias.group(1)
id_video
```

```python
for url in urls:
    patron = r'https://www.youtube.com/watch\?v=(.+)'
    coincidencias = re.search(patron, url)
    if coincidencias:
        id_video = coincidencias.group(1)
        print("Extrayendo transcripción para el video:", id_video)

        # Obtener la transcripción del video
        try:
            transcript_list = YouTubeTranscriptApi.list_transcripts(id_video)
            transcript = transcript_list.find_generated_transcript(['es'])
            texto_concatenado = ' '.join(item['text'] for item in transcript.fetch())

            # Definir la ruta del archivo de destino
            ruta_base = os.path.expanduser("~/Desktop/Redes/transciptor/transcripcion_{}".format(id_video))
            contador = 1
            ruta_archivo = ruta_base + "_" + str(contador) + ".txt"

            # Verificar si el archivo existe y agregar un sufijo numérico si es necesario
            while os.path.exists(ruta_archivo):
                contador += 1
                ruta_archivo = ruta_base + "_" + str(contador) + ".txt"

            # Guardar el texto concatenado en el archivo
            with open(ruta_archivo, 'w', encoding='utf-8') as archivo:
                archivo.write(texto_concatenado)

            print("La transcripción se ha guardado con éxito en:", ruta_archivo)
        except Exception as e:
            print("Error al obtener la transcripción:", str(e))
    else:
        print("URL inválida:", url)
```
